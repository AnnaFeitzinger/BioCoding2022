During the annealing step of PCR, the reaction temprature is lowered to allow primers to bind to single stranded DNA. The temprature must be low enough for the pimers to bind selectively, but not so low that the single-stranded DNA cools and re-forms into a double-stranded helix. 

The formula for calculating the primer melting tempature (Tm) is 

**For primer sequences of 13 bp or less**

*Tm= (wA+xT) * 2 + (yG+zC) * 4*

**For primer sequences of 14 bp or more is**

 *Tm= 64.9 +41*(yG+zC-16.4)/(wA+xT+yG+zC)

Where w,x,y,z are constants that represent the number of nucloetides.

rbcLa rev: 5’- CAGGAAACAGCTATGACGTAAAATCAAGTCCACCRCG-3’
